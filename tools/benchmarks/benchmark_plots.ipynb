```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tf = pd.read_csv('./tf_benchmark_times.csv', names=['L', 'k', 'time']).pivot(columns='k', index='L', values='time')
pt = pd.read_csv('./pt_benchmark_times.csv', names=['L', 'k', 'time']).pivot(columns='k', index='L', values='time')
```

```python
full = pt.join(tf, on='L', lsuffix='_pt', rsuffix='_tf')
```

```python
with plt.style.context('bmh'):
    ax = full.plot(title="PyTorch vs TensorFlow on GPU; 250 epochs; batch size of 32", 
                   figsize=(16,10), style=['k-', 'b-', 'k--', 'b--'])
    ax.set_ylabel("Time Elapsed (s)")
```
